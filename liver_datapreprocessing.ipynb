```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
columns=[
    'Age',
    'Gender',
    'TB',
    'DB',
    'AAP',
    'SGPT',
    'SGOT',
    'TP',
    'ALB',
    'AG Ratio',
    'Disease']
data=pd.read_csv("/Users/vsvsvarma/Desktop/HL_pred/Indian Liver Patient Dataset (ILPD) - Indian Liver Patient Dataset (ILPD).csv")
data.columns=columns
n=578
```

```python
data.index=np.arange(1,n+1)
```

```python
gend=np.empty([578,1])
j=0
for i in data.itertuples():
    if(i.Gender=='Male'):
        gend[j]=1
    else:
        gend[j]=0
    j+=1  
```

```python
data['Gend']=gend
```

```python
data1=data
del data1['Gender']
```

```python
datay=data1['Disease']
datax=data1
del datax['Disease']
del datax['Gend']
datax
```

```python
dis=np.empty([578,1])
j=0
for i in datay:
    if(i==2):
        dis[j]=0
    else:
        dis[j]=1
    j+=1    
dis
```

```python
normalised_data=preprocessing.normalize(datax)
```

```python
norm_data=pd.DataFrame(normalised_data)
columns=[
    'Age',
    'TB',
    'DB',
    'AAP',
    'SGPT',
    'SGOT',
    'TP',
    'ALB',
    'AG Ratio']
norm_data.columns=columns
norm_data['Gender']=gend
norm_data.index=np.arange(1,n+1)
norm_data['Disease']=dis
norm_data
```
